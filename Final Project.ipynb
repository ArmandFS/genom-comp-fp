### Calculate Genome and Codon Statistics & Frequency

```python
def DNAFrequency(A, C, G , T):
    count_A = 0
    count_C = 0
    count_G = 0
    count_T = 0
    noise = 0
    #iterarion for every character
    for char in rec_seq:
        if char == A:
            count_A += 1
        elif char == C:
            count_C += 1
        elif char == G:
            count_G += 1
        elif char == T:
            count_T += 1
        elif char != A or C or G or T:
            noise += 1
    total_A = print("Total DNA A = "+ str(count_A))
    total_C = print("Total DNA C = "+ str(count_C))
    total_G = print("Total DNA G = "+ str(count_G))
    total_T = print("Total DNA T = "+ str(count_T))
    total_noise = print("Total Noise/Error = "+ str(noise))

    print("\nTotal Frequency\n")

    print("Frequency A = " + str(count_A/(len(rec_seq)-noise)))
    print("Frequency C = " + str(count_C/(len(rec_seq)-noise)))
    print("Frequency G = " + str(count_G/(len(rec_seq)-noise)))
    print("Frequency T = " + str(count_T/(len(rec_seq)-noise)))
        
        
#call the dna frequency function
DNAFrequency('A', 'C', 'G', 'T')
```

### Calculate Nucleotide and A-T C-G Density and Plotting them.
### Also added Dimer calculation program

```python
def a_content(seq, start, end):
    window = seq[start:end]
    if len(window) == 0:
        return 0
    a_count = 0
    for char in window:
        if char == 'A':
            a_count += 1
    return a_count / len(window)

def c_content(seq, start, end):
    window = seq[start:end]
    if len(window) == 0:
        return 0
    c_count = 0
    for char in window:
        if char == 'C':
            c_count += 1
    return c_count / len(window)

def g_content(seq, start, end):
    window = seq[start:end]
    if len(window) == 0:
        return 0
    g_count = 0
    for char in window:
        if char == 'G':
            g_count += 1
    return g_count / len(window)

def t_content(seq, start, end):
    window = seq[start:end]
    if len(window) == 0:
        return 0
    t_count = 0
    for char in window:
        if char == 'T':
            t_count += 1
    return t_count / len(window)

def gc_content(seq, start, end):
    window = seq[start:end]
    if len(window) == 0:
        return 0
    gc_count = 0
    for base in window:
        if base in ['G', 'C']:
            gc_count += 1
    return gc_count / len(window)

def at_content(seq, start, end):
    window = seq[start:end]
    if len(window) == 0:
        return 0
    at_count = 0
    for base in window:
        if base in ['A', 'T']:
            at_count += 1
    return at_count / len(window)

a_contents = []
c_contents = []
g_contents = []
t_contents = []
at_contents = []
gc_contents = []

#plotting the sliding window for GC content
def plot(wd):
    a_contents.clear()
    c_contents.clear()
    g_contents.clear()
    t_contents.clear()
    at_contents.clear()
    gc_contents.clear()
    for i in range(0, len(rec_seq) + 1, 5000):
        a_contents.append(a_content(rec_seq, i, i + wd))
        c_contents.append(c_content(rec_seq, i, i + wd))
        g_contents.append(g_content(rec_seq, i, i + wd))
        t_contents.append(t_content(rec_seq, i, i + wd))
        gc_contents.append(gc_content(rec_seq, i, i + wd))
        at_contents.append(at_content(rec_seq, i, i + wd))
    
    #x = np.arange(0, len(a_contents)/10, 0.1)
    x = np.linspace(0, len(rec_seq)/10**6, len(a_contents))

    #plot
    fig, ax = plt.subplots(2, figsize=(10,10))
    ax[0].set_title("Nucleotide Density")
    ax[0].set_xlabel("x10^6")
    ax[0].set_ylabel("Frequency")
    ax[0].plot(x, a_contents, label='A')
    ax[0].plot(x, c_contents, label='C')
    ax[0].plot(x, g_contents, label='G')
    ax[0].plot(x, t_contents, label='T')
    ax[0].set_xlim(0, len(rec_seq)/10**6)
    ax[0].legend()
    ax[0].grid()

    ax[1].set_title("A–T C–G Density")
    ax[1].set_xlabel("x10^6")
    ax[1].set_ylabel("Frequency")
    ax[1].plot(x, gc_contents, label='G-C')
    ax[1].plot(x, at_contents, label='A-T')
    ax[1].set_xlim(0, len(rec_seq)/10**6)
    ax[1].legend()
    ax[1].grid()

    return plt.show()

def dimer():
    AA, AT, AG, AC = 0,0,0,0
    TA, TT, TG, TC = 0,0,0,0
    GA, GT, GG, GC = 0,0,0,0
    CA, CT, CG, CC = 0,0,0,0

    for i in range((len(rec_seq)//2)):
        if rec_seq[i*2] == 'A':
            if rec_seq[i*2+1] == 'A':
                AA +=1
            elif rec_seq[i*2+1] == 'T':
                AT +=1
            elif rec_seq[i*2+1] == 'G':
                AG +=1
            elif rec_seq[i*2+1] == 'C':
                AC +=1

        if rec_seq[i*2] == 'T':
            if rec_seq[i*2+1] == 'A':
                TA += 1
            elif rec_seq[i*2+1] == 'T':
                TT += 1
            elif rec_seq[i*2+1] == 'G':
                TG += 1
            elif rec_seq[i*2+1] == 'C':
                TC += 1
        
        if rec_seq[i*2] == 'G':
            if rec_seq[i*2+1] == 'A':
                GA += 1
            elif rec_seq[i*2+1] == 'T':
                GT += 1
            elif rec_seq[i*2+1] == 'G':
                GG += 1
            elif rec_seq[i*2+1] == 'C':
                GC += 1
        
        if rec_seq[i*2] == 'C':
            if rec_seq[i*2+1] == 'A':
                CA += 1
            elif rec_seq[i*2+1] == 'T':
                CT += 1
            elif rec_seq[i*2+1] == 'G':
                CG += 1
            elif rec_seq[i*2+1] == 'C':
                CC += 1
    
    dAA, dAT, dAG, dAC = AA/(len(rec_seq)/2), AT/(len(rec_seq)/2), AG/(len(rec_seq)/2), AC/(len(rec_seq)/2)
    dTA, dTT, dTG, dTC = TA/(len(rec_seq)/2), TT/(len(rec_seq)/2), TG/(len(rec_seq)/2), TC/(len(rec_seq)/2)
    dGA, dGT, dGG, dGC = GA/(len(rec_seq)/2), GT/(len(rec_seq)/2), GG/(len(rec_seq)/2), GC/(len(rec_seq)/2)
    dCA, dCT, dCG, dCC = CA/(len(rec_seq)/2), CT/(len(rec_seq)/2), CG/(len(rec_seq)/2), CC/(len(rec_seq)/2)

    print("\n\t2-Mer Frequencies")

    print("AA : " + "{:.5f}".format(dAA) + "\tAC : " + "{:.5f}".format(dAC)+"\tAG : "+"{:.5f}".format(dAG)+"\tAT : "+"{:.5f}".format(dAT))
    print("CA : " + "{:.5f}".format(dCA) + "\tCC : " + "{:.5f}".format(dCC)+"\tCG : "+"{:.5f}".format(dCG)+"\tAT : "+"{:.5f}".format(dCT))
    print("GA : " + "{:.5f}".format(dGA) + "\tGC : " + "{:.5f}".format(dGC)+"\tGG : "+"{:.5f}".format(dGG)+"\tGT : "+"{:.5f}".format(dGT))
    print("TA : " + "{:.5f}".format(dTA) + "\tTC : " + "{:.5f}".format(dTC)+"\tTG : "+"{:.5f}".format(dTG)+"\tTT : "+"{:.5f}".format(dTT))
 
# call function
dimer()
wd= widgets.IntText(value= 90000, description='Window Size : ')
widgets.interact_manual(plot, wd=wd)
```

```python
k = int(input("Minimum Lenght:"))
print("Minimum Length:", k)
list_1, list_2, list_3, list_4, list_5, list_6 = [], [], [], [], [], []
orf_list = []
orf_start = []
orf_stop = []
orf_length = []
orf_fram = []
#define the orf function
def orf(k):
    #initialize the frame-specific lists for all six reading frames
    fram_plus1, fram_plus2, fram_plus3 = [], [], []
    fram_min1, fram_min2, fram_min3 = [], [], []
    list_1.clear(), list_2.clear(), list_3.clear(), list_4.clear(), list_5.clear(), list_6.clear()
    
    #the dna is converted into a string and conveted using the upper() function for capitalization
    dna = str(rec_seq.upper())
    fram_plus1.append([dna[i:i + 3] for i in range(0, len(dna), 3)])
    #loop through all the frames here
    for i in range(0,len(fram_plus1),1): 
        start=0
        #this while loop does the looping for each frame for start and stop codons 
        while start <len(fram_plus1[i]): 
            if fram_plus1[i][start]=="ATG":
                for stop in range(start+1,len(fram_plus1[i]),1):
                    if fram_plus1[i][stop]=="TAA" or  fram_plus1[i][stop]=="TAG" or  fram_plus1[i][stop]=="TGA" :
                        test = ''.join(fram_plus1[i][start:stop+1])
                        #retrieve the orf using the append function
                        list_1.append(test) 
                        #keep in mind to avoid all multiple start codons
                        start=stop+1 
                        break
            start+=1
    #print the length of the dna and output of the looping
    count_t = 0
    for i in range(0, len(list_1)):
        if len(list_1[i]) > k:
            count_t += 1
            start_1 = dna.index(list_1[i])+1
            stop_1 = start_1+len(list_1[i])
            length_1 = stop_1-start_1
            print("Frame Type: +1", "Start:", start_1, "Stop:", stop_1, "\nLength:", length_1, "\nORF[", count_t,"]", list_1[i], '\n')

    #+2 frame
    fram_plus2.append([dna[i:i + 3] for i in range(1, len(dna), 3)]) 
    #loop towards all the frames
    for i in range(0,len(fram_plus2),1): 
        start=0
        #looping each frame for start and stop codons 
        while start <len(fram_plus2[i]): 
            if fram_plus2[i][start]=="ATG":
                for stop in range(start+1,len(fram_plus2[i]),1):
                    if fram_plus2[i][stop]=="TAA" or  fram_plus2[i][stop]=="TAG" or  fram_plus2[i][stop]=="TGA" :
                        test = ''.join(fram_plus2[i][start:stop+1])
                        #retrieve the orf using append
                        list_2.append(test) 
                         # avoiding multiple start codons
                        start=stop+1
                        break
            start+=1
    for i in range(0, len(list_2)):
        if len(list_2[i]) > k:
            count_t += 1
            start_2 = dna.index(list_2[i])+1
            stop_2 = start_2+len(list_2[i])
            length_2 = stop_2-start_2
            print("Frame Type: +2", "Start:", start_2, "Stop:", stop_2, "\nLength:", length_2, "\nORF[", count_t,"]", list_2[i], '\n')
    #+3 frame
    fram_plus3.append([dna[i:i + 3] for i in range(2, len(dna), 3)]) 
    #looping all the frames
    for i in range(0,len(fram_plus3),1): 
        start=0
        #looping each frame for start and stop codons
        while start <len(fram_plus3[i]): 
            if fram_plus3[i][start]=="ATG":
                for stop in range(start+1,len(fram_plus3[i]),1):
                    if fram_plus3[i][stop]=="TAA" or  fram_plus3[i][stop]=="TAG" or  fram_plus3[i][stop]=="TGA" :
                        test = ''.join(fram_plus3[i][start:stop+1])
                        #retrieve the orf 
                        list_3.append(test)
                        #avoiding multiple start codons
                        start=stop+1 
                        break
            start+=1
    for i in range(0, len(list_3)):
        if len(list_3[i]) > k:
            count_t += 1
            start_3 = dna.index(list_3[i])+1
            stop_3 = start_3+len(list_3[i])
            length_3 = stop_3-start_3
            print("Frame Type: +3", "Start:", start_3, "Stop:", stop_3, "\nLength:", length_3, "\nORF[", count_t,"]", list_3[i], '\n')

    inv = str(dna[::-1])
    inv = inv.translate(str.maketrans("ATGC","TACG"))
    #-1 frame
    fram_min1.append([inv[i:i + 3] for i in range(0, len(inv), 3)]) 
     #looping all the frames
    for i in range(0,len(fram_min1),1):
        start=0
        #looping each frame for start and stop codons 
        while start <len(fram_min1[i]): 
            if fram_min1[i][start]=="ATG":
                for stop in range(start+1,len(fram_min1[i]),1):
                    if fram_min1[i][stop]=="TAA" or  fram_min1[i][stop]=="TAG" or fram_min1[i][stop]=="TGA" :
                        test = ''.join(fram_min1[i][start:stop+1])
                        #retrieve the orf 
                        list_4.append(test) 
                        #avoiding multiple start codons
                        start=stop+1 
                        break
            start+=1
    for i in range(0, len(list_4)):
        if len(list_4[i]) > k:
            count_t += 1
            start_4 = len(inv)-inv.index(list_4[i])+1
            stop_4 = start_4-len(list_4[i])
            length_4 = start_4-stop_4
            print("Frame Type: -1", "Start:", start_4, "Stop:", stop_4, "\nLength:", length_4, "\nORF[", count_t,"]", list_4[i], '\n')
    #-2 frame
    fram_min2.append([inv[i:i + 3] for i in range(1, len(inv), 3)])
    #looping all the frames
    for i in range(0,len(fram_min2),1):
        start=0
        #looping each frame for start and stop codons 
        while start <len(fram_min2[i]): 
            if fram_min2[i][start]=="ATG":
                for stop in range(start+1,len(fram_min2[i]),1):
                    if fram_min2[i][stop]=="TAA" or  fram_min2[i][stop]=="TAG" or fram_min2[i][stop]=="TGA" :
                        test = ''.join(fram_min2[i][start:stop+1])
                        #retrieve the orf 
                        list_5.append(test)
                        #avoiding multiple start codons 
                        start=stop+1 
                        break
            start+=1
    for i in range(0, len(list_5)):
        if len(list_5[i]) > k:
            count_t += 1
            start_5 = len(inv)-inv.index(list_5[i])+1
            stop_5 = start_5-len(list_5[i])
            length_5 = start_5-stop_5
            print("Frame Type: -2", "Start:", start_5, "Stop:", stop_5, "\nLength:", length_5, "\nORF[", count_t,"]", list_5[i], '\n')
    #-3 frame
    fram_min3.append([inv[i:i + 3] for i in range(2, len(inv), 3)]) 
    #looping all the frames
    for i in range(0,len(fram_min3),1): 
        start=0
        #looping each frame for start and stop codons 
        while start <len(fram_min3[i]): 
            if fram_min3[i][start]=="ATG":
                for stop in range(start+1,len(fram_min3[i]),1):
                    if fram_min3[i][stop]=="TAA" or  fram_min3[i][stop]=="TAG" or fram_min3[i][stop]=="TGA" :
                        test = ''.join(fram_min3[i][start:stop+1])
                        #retrieve the orf 
                        list_6.append(test) 
                        #avoiding multiple start codons
                        start=stop+1 
                        break
            start+=1

    for i in range(0, len(list_6)):
        if len(list_6[i]) > k:
            count_t += 1
            start_6 = len(inv)-inv.index(list_6[i])+1
            stop_6 = start_6-len(list_6[i])
            length_6 = start_6-stop_6
            print("Frame Type: -3", "Start:", start_6, "Stop:", stop_6, "\nLength:", length_6, "\nORF[", count_t,"]", list_6[i], '\n')
    
    print("Total ORFs Found:", count_t)

def orf_count(k):
    dna = str(rec_seq.upper())
    orf_list = []
    
    for frame in range(3):
        for start in range(frame, len(dna)-2, 3):
            codon = dna[start:start+3]
            if codon == "ATG":
                for stop in range(start+3, len(dna), 3):
                    if dna[stop:stop+3] in ["TAA", "TAG", "TGA"]:
                        orf = dna[start:stop+3]
                        if len(orf) > k:
                            orf_list.append((orf, start+1, stop+3, frame))
                        break
    count_t = 0
    for orf, start, stop, frame in orf_list:
        count_t += 1
        print(f"Frame Type: {'+' if frame < 3 else '-'}{frame % 3 + 1} Start: {start} Stop: {stop}\nLength: {len(orf)}\nORF[{count_t}] {orf}\n")
    
    print("Total ORFs Found:", count_t)
    return count_t

    
orf(k)
```
